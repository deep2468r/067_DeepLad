```python
#Declare label as last column in the source file
label = data.iloc[ :, -1 ].values

print(label)
```

```python
# Declaring X as all columns excluding last
X = data.iloc[ :, :-1 ].values
```

```python
# Splitting data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, label, test_size = 0.2, random_state = 67 )
```

```python
# Scaling data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Variabes to calculate sigmoid function
y_pred = []
len_x = len(X_train[0])
w = []
b = 0.2

print(len_x)
```

```python
entries = len(X_train[ :, 0 ])

entries
```

```python
for weights in range(len_x):
    w.append(0)

w
```

```python
# Sigmoid function
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```python
def predict( x, y ):
    lin_reg = np.dot( x, w ) + b
    y_prob = sigmoid( lin_reg )
    y_pred = [1 if i > 0.5 else 0 for i in y_prob]
    return y_pred
```

```python
import math

#Loss function
def loss_func( y, a ):
    J=0;

    for i in range(len(y)):
      if y[i] == 0:
        J += (1-y[i])*math.log(1-a[i])
      else :
        J += (y[i])*math.log(a[i])

    J = J -(2*J)
    return J         
```

```python
dw = []
db = 0
J = 0
alpha = 0.1
loss = []

for x in range(len_x):
    dw.append(0)
```

```python
# Adjusting weights and biases
def gradient_descent( x_train, y_train, w, b, learning_rate, iter ):
  n = len(x_train)
  for i in range(iter):
    lin_reg = np.dot( x_train, w ) + b
    y_prob = sigmoid(lin_reg)

    loss.append( loss_func( y_train, y_prob ) )
    dw = (1/n) * np.dot( x_train.T, y_prob - y_train )
    db = (1/n) * np.sum( y_prob - y_train )

    w = w - learning_rate*dw
    b = b - learning_rate*db
  return ( w, b, loss )

```

```python
# Repeating the process 3000 times
( w, b, loss ) = gradient_descent( X_train, y_train, w, b, alpha, 3000 )
```

```python
# Print weight
print("Adjusted Weights :- \n", w)
```

```python
# Print bias
print("Adjusted bias :- \n", b)
```

```python
# Predicting the label
final_preds = predict( X_test, y_test )
```

```python
# Print actual and predicted values in a table
print("Predicted :- \n", final_preds)
print("\nActual :- \n", y_test)
```

```python
# Calculating accuracy of prediction
from sklearn.metrics import accuracy_score
print( "Accuracy = ", accuracy_score( y_test, final_preds ) )
```
