**1) Create Two numpy array of size 3 X 2 and 2 X 3**

**2) Randomly Initalize that array**

```python
arr1 = np.random.randint( 10, size=(3, 2) )
arr2 = np.random.randint( 10, size=(2, 3) )

print(arr1)
print(arr2)
```

**3) Perform matrix multiplication**

```python
arr_mul = np.dot(arr1, arr2)
#arr_mul = np.matmul(arr1, arr2)

print(arr_mul)
```

**4) Perform elementwise matrix multiplication**

```python
#Transpose of arr2 because both matrix should have same dimensions
t_arr2 = arr2.T

arr_mul_elmnt = np.multiply( arr1, t_arr2 )

print( arr_mul_elmnt )
```

**5) Find mean of first matrix**

```python
arr1_mean = np.mean(arr1) 

print( arr1_mean )
```
